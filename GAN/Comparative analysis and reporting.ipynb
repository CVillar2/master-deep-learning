### Conclusion

Lets talk now a bit about the practise the obtained results and some of the problems found.

In one hand, this practise allowed us to get familiar with VAEs and GANs and how they work, in the other hand, the fact that they are so heavily time consuming and computing resources consuming is a problem. We can say that this fact was a bottleneck as avery time we wanted to try something we had to wait a lot of time or it simply did not work because the CPU could not run it properly. We also had problems with the version of the libraries and the kernel crashing because of it but were able to solve it quite fast.

If we take a look into the results we can see that we are going in the right way, but the fact that the models need so much time to train did not allow us to get good results, we would need a computer much more powerful and more time to let it train. 

Comparing now the results of the VAEs and the GANs, the faces generated by the VAEs are closer to look like a real face than the ones generated by the GANs, also the VAEs required less CPU and time resources. In this comparison it is important to take into account that the VAEs try to generate known faces while GANs try to generate new faces.
